## Part 4: Extract Features

Create features from the accelerometer data. Features should summarize the motion in a way that distinguishes steady from shaking.

```python
def extract_features_from_window(accel_x, accel_y, accel_z):
    """Extract features from a window of accelerometer data.
    
    Parameters:
    - accel_x, accel_y, accel_z: arrays of acceleration values
    
    Returns:
    - features: dictionary with extracted features
    """
    
    features = {}
    
    # TODO: Extract features. Try these:
    
    # 1. Magnitude (total acceleration)
    magnitude = np.sqrt(accel_x**2 + accel_y**2 + accel_z**2)
    
    # 2. Statistical features for each axis
    for axis, data in [('x', accel_x), ('y', accel_y), ('z', accel_z)]:
        features[f'mean_{axis}'] = np.mean(data)
        features[f'std_{axis}'] = np.std(data)
        features[f'var_{axis}'] = np.var(data)
        features[f'min_{axis}'] = np.min(data)
        features[f'max_{axis}'] = np.max(data)
        features[f'range_{axis}'] = np.max(data) - np.min(data)
    
    # 3. Magnitude statistics
    features['mean_magnitude'] = np.mean(magnitude)
    features['std_magnitude'] = np.std(magnitude)
    features['var_magnitude'] = np.var(magnitude)
    
    # 4. Additional features to try:
    # features['rms_x'] = np.sqrt(np.mean(accel_x**2))
    # features['rms_y'] = np.sqrt(np.mean(accel_y**2))
    # features['rms_z'] = np.sqrt(np.mean(accel_z**2))
    # features['skewness_x'] = stats.skew(accel_x)
    # features['skewness_y'] = stats.skew(accel_y)
    # features['skewness_z'] = stats.skew(accel_z)
    # features['kurtosis_x'] = stats.kurtosis(accel_x)
    
    return features

print('Feature extraction function defined')
print('Features to extract:')
print('- Mean, std, variance for each axis')
print('- Min, max, range for each axis')
print('- Magnitude statistics')
```
